```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import image
```

```python
model = tf.keras.models.load_model("../artifacts/model.h5")
```

```python
model.summary()
```

```python
img = image.imread("../imagem.png")
```

```python
img.shape
```

```python
plt.imshow(img, cmap="gray");
```

```python
img_resize = tf.image.resize(img, (28, 28), method="area").numpy()
img_resize = tf.image.rgb_to_grayscale(img_resize).numpy()
```

```python
plt.imshow(img_resize, cmap="gray");
```

```python
img_resize.shape
```

```python
plt.imshow(img_resize, cmap="gray");
```

```python
np.expand_dims(img_resize, 0)
```

```python
pred = np.argmax(model.predict(np.expand_dims(img_resize, 0)))
```

```python
pred
```

```python
plt.imshow(img_resize, cmap="gray");
plt.title(f"Label: 3   Predict: {pred}");
```
